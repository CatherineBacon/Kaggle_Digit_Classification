```python
# import libraries
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
# read in data
dataset = np.genfromtxt(open('train.csv', 'r'),
                        delimiter=',', dtype='f8')[1:]
label = [x[0] for x in dataset]
pixels = [x[1:] for x in dataset]
test_pixels = np.genfromtxt(open('test.csv', 'r'),
                     delimiter =',', dtype='f8')[1:]
```

```python
# take pixel columns and make into an array of 8-bits
# resutling array will be 1-dimentional, length of 784
# pixel intensities are integers from 0 to 255
pixels = np.array(pixels, dtype='uint8')
```

```python
# plot
for index, row in enumerate(pixels[:4]):
    # reshape into 28x28 array
    image = np.reshape(row,(28,28))
    # plot
    plt.axis('off')
    plt.title('Label is {label}'.format(label=label[index]))
    plt.imshow(image, cmap='gray')
    plt.show()

```

```python
# now look at test set with predicted labels
test_labels = np.genfromtxt(open('results-v3.csv', 'r'),
                     delimiter =',', dtype='f8')[1:,1:]
```

```python
test_pixels = np.array(test_pixels, dtype='uint8')
```

```python
for index, row in enumerate(test_pixels[:4]):
    # reshape into 28x28 array
    image = np.reshape(row,(28,28))
    # plot
    plt.axis('off')
    plt.title('Label is {label}'.format(label=int(test_labels[index][0])))
    plt.imshow(image, cmap='gray')
    plt.show()
```
